<h3> EDA Step 1 (Data Collection) and Step 2 (Data Preparation and Understanding Variables) - Code </h3>
<p> Ref.[3] </p>

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset into a dataframe
df = pd.read_csv('BankRecords.csv')

# Display the first few rows
print("First few rows of the dataset:")
print(df.head(), "\n")

# Output of dimensions of the dataset
print("Dimensions of the dataset:", df.shape, "\n")

# Output of missing values
print("Missing values:\n", df.isnull().sum(), "\n")

# Data types of each feature
print("Data types:\n", df.dtypes, "\n")

# Data columns of full dataset
print("Columns:\n", df.columns, "\n")

# Numerical features statistics
print("Statistics (summary):\n", df.describe(), "\n")

# Getting duplicated rows
duplicated_rows = df[df.duplicated(keep=False)]
# Test
# print("Duplicated rows:\n", duplicated_rows)

# Count duplicated rows and output
duplicated_counts = df.duplicated().sum()
print("Number of duplicated rows:", duplicated_counts)

# Identify outliers
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)
outliers_count = outliers.sum()
print("Number of rows with outliers:", outliers_count)

# Create a bar plot to show outliers visualization
plt.figure(figsize=(8, 6))
plt.bar(["Rows with Outliers"], [outliers_count], color='orange')
plt.xlabel("Data Points")
plt.ylabel("Count")
plt.title("Count of Rows with Outliers")
plt.show()

# Check for unique values in categorical columns - data validation
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    unique_values = df[column].unique()
    print(f"Unique values in {column}:", unique_values, "\n")

# Transpose values for a better overview
print("Transpose index and columns:\n", df.describe().T)
```

<h3> EDA Step 3 (Data Preparation) - Code </h3>
<p> Ref.[4] & [5] </p>
<p> Addressing data quality issues identified in the previous step. </p>
<b> Overview </b>
<p> - Correct or remove rows with negative experience values, meaning to handle negative experience values. </p>
<p> - Ensure that other columns are free from anomalities, such as verify and clean data. </p>
<p> - Apply transformations, encode, and standardize. </p>
<b> How did I handle negative experience values:</b>
<p> Identified rows with negative experience values, then replaced negative values with the median experience of non-negative values to correct anomalies. </p>
<b> How did I encoded the categorical variables:</b>
<p> By the use of 'LabelEncoder', to convert categorical features into numerical format. </p>
<b> How did I performed feature engineering by standardazing the numerical features:  </b>
<p> I applied 'StandardScaler' to ensure that numerical features would have a mean of 0 and a standard deviation of 1. </p>
<b> Plotting:</b>
<p> After all the fixes performed above, there was the need to visualize some informations such as the distribution of experience and income after cleaning. Plotted as well a scatter plot to show the relationship between experience and income after cleaning. </p>
<b> About update on findings and research: </b>
<p> A deeper research was performed to analyse into details the coorelation between such features, and to assure that the features chosen were the correct one for such analyse and outcome (customer income). And a necessary in-depth coorellation analysis will be needed.<p> 
<p> <b> Experience data:</b>there was found negative values, and they were corrected, which resultted in a more accurate representation of customers' work experience. </p>
<p> <b> Income distribution:</b> The income distribution was still right-skewed, and there was a need to considerate it better during the model training and inpacts in such. </p>
<p> <b> Experience-Independednt Income:</b> The scatter plot continued to show a lack of strong coorelation between experience and income, with need to explore other features. </p>
<p> With all the above found and addressed data quality issues, it is possible to proceed with the next step, which is identifying coorelated variables to predict customer income, work with the target variable and it's visualization. Check coorelations with categorical features vs target variable and numerical feature vs target variable. </p>

```python
# Importing necessary libraries 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Handling negative experience values
# Replaced negative experience values with the median experience
median_experience = df[df['Experience(Years)'] >= 0]['Experience(Years)'].median()
df.loc[df['Experience(Years)'] < 0, 'Experience(Years)'] = median_experience

# Encode categorical variables
label_encoders = {}
categorical_columns = ['Personal Loan', 'Securities Account', 'CD Account', 'Online Banking', 'CreditCard', 'Education']
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Standardize numerical features
scaler = StandardScaler()
numerical_columns = ['Age', 'Experience(Years)', 'Sort Code', 'Credit Score', 'Mortgage(Thousands\'s)']
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Verification of the cleaning process
print("Summary statistics after cleaning:")
print(df.describe())

# Visualizing the cleaned Experience (Years) data
plt.figure(figsize=(10, 5))
sns.histplot(df['Experience(Years)'], kde=True)
plt.title('Distribution of Experience (Years) after Cleaning')
plt.show()

# Visualizing the cleaned Income data
plt.figure(figsize=(10, 5))
sns.histplot(df['Income(Thousands\'s)'], kde=True)
plt.title('Distribution of Income (Thousands) after Cleaning')
plt.show()

# Visualizing the relationship between Experience and Income after cleaning
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Experience(Years)'], y=df['Income(Thousands\'s)'])
plt.title('Experience vs. Income after Cleaning')
plt.show()

# Splitting the data into features and target variable
X = df.drop(columns=['Income(Thousands\'s)', 'ID'])
y = df['Income(Thousands\'s)']
```

<h3> EDA Step 3 (Data Preparation) - Findings </h3>
<p> Ref.[6]</p>
<b> Summary Statistics after cleaning </b>
<p> It is possible to have a comprehensive summary of the statistics for some feature, see more details over the first table provided on the above outcome. </p>
<b> Key findings </b>
<p> <b> Experience in years:</b> The mean of the standardized experience is near 0, with a standard deviation of 1. The values range from -2.89 to 5.64 after the standardization process, which shows a wider spread, compared to age. </p> 
<p> <b> Income (thousands's):</b> The mean income is 73.77 thousand dollars, with a standard deviation of 46.03 thousand dollars. It is also possible to notice that the income distribution is positively skewed, with a maximum income of 224 thousand dollars. </p>
<b> About Visualizations: </b>
<p> <b> To conclude:</b> The above details, shows distinct pattersn in the data: The first histogram shows a highly concentrated distribution for personal loands, the second indicates a right-skewed distribution for income, and the scatter plot shows a dispersed relationship between credit score and income, suggesting weak coorelation. Those insights will help to guide us over the next step, to understand customer profiles more effectively. </p>
